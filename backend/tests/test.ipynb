```python
import random

def remplir_grille(grille):

    # Chercher une case vide
    for i in range(9):
        for j in range(9):
            if grille[i][j] == 0:
                cases = (i, j)
                break
        else:
            continue
        break
    else:
        # Plus aucune case vide → grille complète
        return True

    numbers = list(range(1, 10))
    random.shuffle(numbers)

    for n in numbers:

        # Vérifier ligne
        if n in grille[cases[0]]:
            continue

        # Vérifier colonne
        if n in [grille[x][cases[1]] for x in range(9)]:
            continue

        # Vérifier bloc 3x3
        bloc_i = (cases[0] // 3) * 3
        bloc_j = (cases[1] // 3) * 3

        if n in [grille[bloc_i + k][bloc_j + l] for k in range(3) for l in range(3)]:
            continue

        # Placer le nombre
        grille[cases[0]][cases[1]] = n

        # Appel récursif
        if remplir_grille(grille):
            return True

        # Backtrack
        grille[cases[0]][cases[1]] = 0

    return False



grille = [[0 for _ in range(9)] for _ in range(9)]
remplir_grille(grille)
print(grille)
    
    
```

fonction remplir(grille):

    si toutes les cases sont remplies:
        retourner VRAI

    choisir une case vide

    mélanger les chiffres 1 à 9 au hasard

    pour chaque chiffre dans 1 à 9:
        si chiffre est valide dans la case:
            placer le chiffre

            si remplir(grille) retourne VRAI:
                retourner VRAI

            sinon:
                effacer la case  (backtrack)

    retourner FAUX
